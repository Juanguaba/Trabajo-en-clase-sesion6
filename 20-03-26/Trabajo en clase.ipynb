```python
import pandas as pd
import numpy as np
import unicodedata

```

```python
url = r"./chat_bot.xlsx"
```

```python
dataset = pd.read_excel("chat_bot.xlsx")
```

```python
dataset
```

```python
print("NovaBot: ¡Hola! Soy tu asistente virtual")
print("Escribe 'salir' para terminar\n")

def normalizar(texto):
    
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    
    texto = texto.replace('¿', '').replace('?', '').replace('¡', '').replace('!', '')
    return texto.lower().strip()

while True:

    usuario = input("Tú: ").lower()

    if usuario == "salir":
        print("NovaBot: ¡Adiós! Que tengas un buen día.")
        break


    respuesta = dataset[dataset["Pregunta"].apply(normalizar) == normalizar(usuario)]


    if not respuesta.empty:
        print("NovaBot:", respuesta["Respuesta"].values[0])
    else:
        print("NovaBot: Lo siento, no entiendo tu pregunta.")
    

    
```
